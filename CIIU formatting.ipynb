```python
import pandas as pd
import unidecode
```

```python
df = pd.read_csv('./data/disagg_struct_refs/sources/ciiu4_v2.csv', header = None, usecols=range(5))
df.columns = ['CIIU ID CI0', 'CIIU ID CI1', 'CIIU ID CI2', 'CIIU ID CI4', 'CIIU label']

# ciiu_0 = df.where(df['CIIU ID CI1'].str.len() > 5, np.nan)[['CIIU ID CI1', 'CIIU label']].dropna()
# ciiu_0.columns = ['CIIU ID CI0', 'CIIU label CI0']

# ciiu_1 = df.where(df['CIIU ID CI1'].str.len() < 5, np.nan)[['CIIU ID CI1', 'CIIU label']].dropna()
# ciiu_1.columns = ['CIIU ID CI1', 'CIIU label CI1']

ciiu_0 = df[['CIIU ID CI0', 'CIIU label']].dropna()
ciiu_0.columns = ['CIIU ID CI0', 'CIIU label CI0']

ciiu_1 = df[['CIIU ID CI1', 'CIIU label']].dropna()
ciiu_1.columns = ['CIIU ID CI1', 'CIIU label CI1']

ciiu_2 = df[['CIIU ID CI2', 'CIIU label']].dropna()
ciiu_2.columns = ['CIIU ID CI2', 'CIIU label CI2']

ciiu_4 = df[['CIIU ID CI4', 'CIIU label']].dropna()
ciiu_4.columns = ['CIIU ID CI4', 'CIIU label CI4']

df_ = df.fillna(method = 'ffill').loc[df['CIIU ID CI4'].notnull()]

df_ = df_.merge(ciiu_0).merge(ciiu_1).merge(ciiu_2).rename({'CIIU label': 'CIIU label CI4'}, axis = 1)
df_[[col for col in df_.columns if 'label' in col]]

df_['CIIU label CI0'] = df_['CIIU label CI0'].str.capitalize()
df_['CIIU ID CI0'] = df_['CIIU ID CI0'].str.rstrip().str[-1]

CIIU = df_

CIIU.to_csv('CIIU.csv')
```
