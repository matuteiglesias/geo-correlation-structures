```python
df1 = pd.read_csv('./data/BR/rankings-location-international-trade.csv')
df2 = pd.read_csv('./data/BR/secex-2016-microregions.csv') # Many fewer...
# df2.shape

def format_col(series):
    return series.str.replace('USD ', '').str.replace('B', 'e9').str.replace('M', 'e6').str.replace('k', 'e3').astype(float)


df1['EST'] = df1['Location'].str[-2:]

df1['Imports'] = format_col(df1['Imports'])
df1['Exports'] = format_col(df1['Exports'])

df2['EST'] = df2['BRA ID'].str[1:3]
```

```python
df1.sort_values(['EST', 'Exports'])[['EST','IBGE ID','Location','Exports']]
```
